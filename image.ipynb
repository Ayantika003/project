```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('/content/images.jfif')
```

```python
cv2_imshow(image)
```

```python
image_resized = cv2.resize(image,None,fx= 0.6,fy=0.6)
```

```python
cv2_imshow(image_resized)
```

 Clear the impurities  --



the impurities may be colour variation due to which filters may lag

```python
image_cleared = cv2.medianBlur(image_resized,3)
image_cleared = cv2.medianBlur(image_resized,3)
image_cleared = cv2.medianBlur(image_resized,3)

image_cleared = cv2.edgePreservingFilter(image_cleared,sigma_s= 5)
```

```python
cv2_imshow(image_cleared)
```

APPLY FILTERING :-

```python
image_filtered = cv2.bilateralFilter(image_cleared,3,10,5)

for i in range(2):
  image_filtered = cv2.bilateralFilter(image_filtered,3,20,10)
for i in range(3):
  image_filtered = cv2.bilateralFilter(image_filtered,5,30,10)
```

```python
cv2_imshow(image_filtered)
```

Tune the Art

```python
# sharping the image using addweighted()
guassian_mask = cv2.GaussianBlur(image_filtered,(7,7),2)
image_sharp = cv2.addWeighted(image_filtered,1.5,guassian_mask,-0.5,0)
```

```python
cv2_imshow(image_sharp)
```
