# **Drawing The Shape**

```python
image = np.zeros((512,512,3))
```

```python
# rectangle
rec_img = cv2.rectangle(image,pt1=(100,100),pt2=(300,300),color=(255,0,0),thickness=5)
```

```python
cv2_imshow(rec_img)
```

```python
# circle
circle = cv2.circle(image,center=(100,400),radius = 50,color=(0,0,255),thickness=-1)
```

```python
cv2_imshow(circle)
```

```python
#line
line = cv2.line(image,pt1=(0,0),pt2=(512,512),color=(0,255,0),thickness=2)
cv2_imshow(line)
```

```python
# Text
line = cv2.putText(image,org=(200,400),fontScale=4,color=(0,255,255),thickness=5,lineType=cv2.LINE_AA,text="Hello",fontFace=cv2.FONT_ITALIC)
cv2_imshow(line)
```
